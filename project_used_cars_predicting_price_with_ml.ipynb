## Dealing with missing values

```python
import numpy as np
df=df.replace('?',np.nan)
```

```python
df.isnull().sum()
```

```python
#filling Nan values with median
df[['Normalized_losses','Bore',
    'Price','Stroke','Horsepower','Peak-rpm']]=df[['Normalized_losses','Bore','Price','Stroke','Horsepower','Peak-rpm']].fillna(df[['Normalized_losses','Bore','Price','Stroke','Horsepower','Peak-rpm']].median())
```

```python
#recasting numerical columns to int and float type
df[['Normalized_losses','Price','Horsepower','Peak-rpm']]=df[['Normalized_losses','Price','Horsepower','Peak-rpm']].astype(int)
```

```python
df[['Bore','Stroke']]=df[['Bore','Stroke']].astype(float)
```

```python
#data formatting
#convert mpg to l/100km
df['City-mpg']=(235/df['City-mpg']).round(1)
```

```python
df.rename(columns={'City-mpg':'City-Litre/100km'},inplace=True)
```

```python
df.head(2)
```

```python
#data normalization
#df.Length.max()
df['Length']=df['Length']/208.1
#df.Width.max()
df['Height']=df['Height']/59.8
df['Width']=df['Width']/72.3
```

```python
df[['Height','Length','Width']].head()
```

```python
#binning
bins=np.linspace((df['Compression-ratio'].min()),(df['Compression-ratio'].max()),4)
group=['low','medium','high']
df['Compression-ratio_binned']=pd.cut(df['Compression-ratio'],bins,labels=group,include_lowest=True)
```

```python
df['Compression-ratio_binned'].unique()
```

```python
#turning categorical variables into numerical with one-hot encoding
dummies=pd.get_dummies(df['Fuel-type'])
```

```python
pd.concat([df,dummies],axis=1)
```
